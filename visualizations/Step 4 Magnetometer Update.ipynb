```python
import numpy as np

import matplotlib.pyplot as plt

%matplotlib inline 
%config InlineBackend.figure_format = 'retina'
```

```python
# time, Quad.MagYaw, Quad.Est.Yaw, Quad.Yaw
no_code_graph1 = np.loadtxt('./data/scenario4/NoCodeGraph1.txt',delimiter=',',dtype='Float64',skiprows=1)
# time, Quad.Est.E.Yaw
no_code_graph2 = np.loadtxt('./data/scenario4/NoCodeGraph2.txt',delimiter=',',dtype='Float64',skiprows=1)
```

```python
def plotYawError(data, title):
    """
    Plots the Yaw error
    """
    fig, ax = plt.subplots(figsize=[20,10])
    time = data[:, 0]
    onces = np.ones(len(time))
    ax.plot(time, data[:,1], 'y')
    ax.plot(time, -.15*onces, 'r--')
    ax.plot(time, .15*onces, 'r--')
    ax.set_xlabel('Time [s]')
    ax.set_ylabel('Yaw Error [rad]')
    ax.grid()
    plt.suptitle(title, fontsize=24)

plotYawError(no_code_graph2, 'Widthout mofifications')
```
